```python
import pandas as pd
```

```python
df=pd.read_csv("/content/IMDB_Modified_Data - modified_data.csv")
df
```

```python
print(df.columns)

```

```python
# Calculate the median of the 'Age' column
median_age = df['Age '].median()

# Replace NaN values with the median in the 'Age' column
df['Age '].fillna(median_age, inplace=True)

print(df)
```

```python
# Replace '#VALUE!' with NaN in the 'Duration' column
df['Duration'].replace('#VALUE!', pd.NA, inplace=True)

# Remove rows with NaN values in the 'Duration' column
df.dropna(subset=['Duration'], inplace=True)

print(df)
```

```python
# Calculate the median of the 'Age' column
rating = df['Rating '].median()

# Replace NaN values with the median in the 'Age' column
df['Rating '].fillna(rating, inplace=True)

print(df)
```

```python
# Calculate the median of the 'Age' column
gross_profit = df['Gross Profit'].median()

# Replace NaN values with the median in the 'Age' column
df['Gross Profit'].fillna(gross_profit, inplace=True)

print(df)
```

```python
# Calculate the median of the 'Age' column
gross_profit = df['Gross Profit'].median()

# Replace NaN values with the median in the 'Age' column
df['Gross Profit'].fillna(gross_profit, inplace=True)

print(df)
```

```python
# Remove NaN values from the 'votes' column
df.dropna(subset=['Votes '], inplace=True)

print(df)
```

```python
df['Votes '] = df['Votes '].str.replace(r'\d+,\d+ \| Gross: \$\d+\.\d+M', '', regex=True)

print(df)
```

```python
df.dropna(subset=['Votes '], inplace=True)

print(df)
```

```python
df.to_csv('modified_data.csv', index=False)
```
